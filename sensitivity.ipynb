# Sensitivity Tests: initial spacing and initial volume transport

Tests in three approaches:

1. Drifter to drifter: compare drifter tracks directly using $D$ (relative dispersion) and $s$ (trajectory score). Actually, I don't think I can do any relative test on this because drifters are starting at different locations in most cases, outside of the cases in which only the number of drifters at a location is changed for transport purposes.
1. Cloud: compare cloud statistics of drifters using first, second, and third moments. Calculate the moments in time, averaged over all drifters, for each simulation and look for convergence to a consistent result with increased resolution. This is expected to matter more on the initial drifter separation than the number of drifters initialized per grid cell.
1. Transport: compare the map of transport between two simulations in time. At the final time in the simulation (the only time we have transport info, actually), find the norm of the transport maps. With increasing resolution in grid spacing and especially drifters representing grid cells, this residual should decrease at some point.

Note: Use a consistent nsteps between the simulations.

```python
seed = 'A'

fig = figure(figsize(20,8))
ax1 = fig.add_subplot(111)
# ax2 = ax1.twinx()

diffnorm = np.zeros((dxs.size,Vs.size))*np.nan
ndrifters = np.zeros((dxs.size,Vs.size))*np.nan
# loop through Vs
for j, V in enumerate(Vs):
    
    # Loop through dxs
    for i, dx in enumerate(dxs):
        
        File = 'calcs/' + seed + '_dx' + str(dx) + '_V' + str(V) + 'diffnorm.npz'
        d = np.load(File)
        diffnorm[i,j] = d['diffnorm']
        d.close()
        
        File = 'tracks/' + seed + '_dx' + str(dx) + '_V' + str(V) + '.nc'
        d = netCDF.Dataset(File)
        ndrifters[i,j] = d.variables['lonp'][:,0].size
        d.close()
        
cs = ax1.contour(dxs, Vs, diffnorm.T, np.arange(2500,100000,2500), colors='0.1', linewidths=1)
ax1.clabel(cs, np.arange(2500,100000,7500), inline_spacing=10, inline=1, fmt='%.0f')#, fontsize=10)

cs = ax1.contour(dxs, Vs, ndrifters.T, np.arange(3000,36000,3000), colors='r', linewidths=5, linestyles='-', zorder=1, alpha=0.5)
ax1.clabel(cs, np.arange(3000,36000,6000), inline_spacing=20, inline=1, fmt='%.0f', fontsize=20)

ax1.set_xlabel('Initial drifter separation [m]')
ax1.set_ylabel('Max volume represented by a drifter [m$^3$s$^{-1}$]')
ax1.set_xlim(250,2000)

# Legend
text(1500, 95, 'Norm of transport differences', backgroundcolor='w', color='0.1', fontsize=20)
text(1500, 90, 'Number of drifters', backgroundcolor='w', color='r', alpha=0.7, fontsize=20)

# # Note location of interest
# text(1000, 95, '$3000$', color='r', alpha='0.5')

savefig('figures/' + seed + 'transport.png', bbox_inches='tight')
savefig('figures/' + seed + 'transport.pdf', bbox_inches='tight')

```
